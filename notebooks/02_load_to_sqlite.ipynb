```python
import sqlite3
import os

sql_path = os.path.join("..", "sql", "01_create_tables.sql")
db_path = os.path.join("..", "data", "processed", "marketplace.db")

conn = sqlite3.connect(db_path)

with open(sql_path, "r", encoding="utf-8") as f:
    create_table_sql = f.read()

conn.executescript(create_table_sql)
conn.close()

print("✅ Table created successfully in:", db_path)
```

```python
import pandas as pd
import sqlite3
import os

csv_path = os.path.join("..", "data", "processed", "online_retail_cleaned.csv")
db_path = os.path.join("..", "data", "processed", "marketplace.db")

df = pd.read_csv(csv_path)

conn = sqlite3.connect(db_path)
df.to_sql("marketplace_orders", conn, if_exists="append", index=False)
conn.close()

print("✅ Data loaded into marketplace_orders:", len(df), "rows")
```
